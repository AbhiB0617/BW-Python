```python
#2. Write a Python program to read first n lines of a file.

n = int(input("Enter the number of lines to read: "))

with open("Newwfile.txt", "r") as file:
    for i in range(n):
        line = file.readline()
        if not line:  # Stop if file has fewer lines
            break
        print(line.strip())
```

```python
#5. Write a Python program to read a file line by line and store it into a list.

with open('Newwfile.txt', 'r') as file:
    lines = []
    for line in file:
        lines.append(line.strip()) # strip() removes newline characters
print(lines)
```

```python
# 6. Write a Python program to read a file line by line store it into a variable.

with open('Newwfile.txt', 'r') as file:
    x = " "
    for line in file:
        x += line   # append each line to the variable
print(x)
```

```python
#7. Write a python program to find the longest words.

with open('Newwfile.txt', 'r') as file:
    words = file.read().split()

max_length = max(len(i) for i in words)
longest_words = [i for i in words if len(i) == max_length]

print(longest_words)
```

```python
# 9. Write a Python program to count the frequency of words in a file.

with open('Newwfile.txt', 'r') as file:
    text = file.read()
words = text.split()
word_count = {} #dictionary to store word frequencies

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

for word, count in word_count.items():
    print(word, ":", count)

```

```python
# 12. Write a Python program to copy the contents of a file to another file .

source_file = 'myfile.txt'
destination_file = 'copied_myfile.txt'

with open(source_file, 'r') as S, open(destination_file, 'w') as D:
    for line in S:
        D.write(line)

print("File copied successfully!")
```

```python
# 14. Write a Python program to assess if a file is closed or not.

file = open('Newwfile.txt', 'r')

try:
    file.readline()
    print("File is open.")
except ValueError:
    # If file is closed, this error is raised
    print("File is closed.")

file.close()

# Check again
try:
    file.readline()
    print("File is open.")
except ValueError:
    print("File is closed.")
```

```python
# 15.Write a Python program that takes a text file as input and returns the number of words of a given text file.

file_name = input("Enter the file name: ")

try:
    with open(file_name, 'r') as file:
        text = file.read()
        words = text.split()  # split text into words
        word_count = len(words)

    print(f"Number of words in '{file_name}': {word_count}")

except Error :
        print(f"File '{file_name}' not found.")
```
